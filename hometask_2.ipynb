```python
def kMeans(matrix, k, iter, eps):
    centers = np.array(matrix[:k])
    
    s = 0
    a = 0
    b = 1
    labels = getLabels(matrix, centers)

    while(abs(b - a) > eps):
        a = b
        
        for j in range(k):
            centers[j] = np.mean(matrix[labels == j], axis = 0)
            
        labels = getLabels(matrix, centers)
        b = np.mean(np.linalg.norm(matrix - centers[labels], axis = 1))
        
        s += 1
        if s == iter:
            break
        
    return [labels, b]

def getLabels(matrix, centers):
    return np.argmin(np.linalg.norm(matrix - centers[:, None], axis = 2), axis = 0)
```

```python
import numpy as np
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/sharmaroshan/Clustering-of-Mall-Customers/master/Mall_Customers.csv'
data = np.loadtxt(url, delimiter=',', dtype=object, skiprows=1)

data[:, 1] = data[:, 1] == 'Male'
data = data.astype(float)
```

```python
plt.scatter(data[:, 2], data[:, 3])

plt.xlabel('Age')
plt.ylabel('Annual Income')
plt.show()
```

```python
arr = []
for k in range (1, 10):
    matrix = data[:, [2, 3]]
    ret = kMeans(matrix, k, 100, 1e-5)
    arr.append(ret[1])
    
plt.plot(arr)
```

```python
k = 3
matrix = data[:, [2, 3]]

ret = kMeans(matrix, k, 100, 1e-5)
labels = ret[0]

for i in range(k):
    tempMatr = matrix[labels == i]
    plt.scatter(*tempMatr.T)
```
